## Best Model Comparison

```python
best_model_df = pd.read_csv('/content/drive/MyDrive/ECSE_552/Project/Submission/Test_Results/Best_Model_Comparison.csv')
best_model_df
```

```python
def grouped_barplot(df, cat,subcat, val , err):
  u = df[cat].unique()
  x = np.arange(len(u))
  subx = df[subcat].unique()
  offsets = (np.arange(len(subx))-np.arange(len(subx)).mean())/(len(subx)+1.)
  width= np.diff(offsets).mean()
  colormap = ['#003f5c', '#58508d', '#bc5090', '#ff6361', '#ffa600']
  for i,gr in enumerate(subx):
      dfg = df[df[subcat] == gr]
      plt.bar(x+offsets[i], dfg[val].values, width=width, 
              label="{}".format(gr), yerr=dfg[err].values, color=colormap[i])
  # plt.xlabel(cat)
  plt.plot(0.155, 1.1, marker='*', color='black')
  plt.plot(0.325, 1.1, marker='*', color='black')
  plt.plot(1, 3.4, marker='*', color='black')
  plt.plot(2, 3.6, marker='*', color='black')
  plt.ylabel('MAPE(%)')
  plt.xticks(x, u)
  plt.ylim((0, max(df[val])*1.3))
  plt.legend()
  plt.show()

cat = "Type"
subcat = "Model"
val = "MAPE AVG"
err = "MAPE STD"
grouped_barplot(best_model_df, cat, subcat, val, err )
```

```python
def grouped_barplot(df, cat,subcat, val , err):
  u = df[cat].unique()
  x = np.arange(len(u))
  subx = df[subcat].unique()
  offsets = (np.arange(len(subx))-np.arange(len(subx)).mean())/(len(subx)+1.)
  width= np.diff(offsets).mean()
  # colormap = ['#813d32', '#ef852f']
  colormap = ['#051e3e', '#851e3e']
  
  for i,gr in enumerate(subx):
      dfg = df[df[subcat] == gr]
      plt.bar(x+offsets[i], dfg[val].values, width=width, 
              label="{}".format(gr), yerr=dfg[err].values, color=colormap[i], alpha=1)
      
  plt.plot(-0.16, 3.75, marker='*', color='black')
  plt.plot(0.83, 3.3, marker='*', color='black')
  plt.ylabel('MAPE(%)')
  plt.xticks(x, u)
  plt.ylim((0, max(df[val])*1.3))
  plt.legend()
  plt.title('Effect of Feature Shifting on 24-Hour Sequential Predictions', y=1.05)
  plt.show()

cat = "Model"
subcat = "Shift"
val = "MAPE AVG"
err = "MAPE STD"
grouped_barplot(Seq_Shift_df, cat, subcat, val, err )
```

# Feature Importance

```python
FeaImp_df = pd.read_csv('/content/drive/MyDrive/ECSE_552/Project/Submission/Test_Results/feature_imp.csv')
FeaImp_df.columns.values[0] = 'Features'
FeaImp_df
```

```python
ind = np.arange(len(FeaImp_df))
width = 0.4

fig, ax = plt.subplots()

ax.barh(ind + width, FeaImp_df['1h, shift'], width, color='#813d32', label='Shift')
ax.barh(ind, FeaImp_df['1h, no shift'], width, color='#ef852f', label='No Shift')

ax.set(yticks=ind + width, yticklabels=FeaImp_df.Features, ylim=[2*width - 1, len(FeaImp_df)])
ax.set_xlim(0, 0.11)
ax.legend(loc='upper right')

plt.gca().invert_yaxis()
plt.show()
```
